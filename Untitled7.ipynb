```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/content/moviesfinal.csv')
print(data)
```

```python
encoder = LabelEncoder()
data["Genre_encoded"] = encoder.fit_transform(data["genre"])
print(data)
```

```python
similarity = cosine_similarity(data[["Genre_encoded"]])
```

```python
def Recommended_movies(movie_name):
    # Convert both the input movie name and the 'movie' column to lowercase for case-insensitive comparison
    movie_index = data[data["movie"].str.lower() == movie_name.lower()].index[0]
    similarity_scores = list(enumerate(similarity[movie_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    print("Recommended Movies:")
    for i in similarity_scores [1:4]:
        print(data.iloc[i[0]]["movie"])
```

```python
Recommended_movies("Avengers: Endgame")
```
